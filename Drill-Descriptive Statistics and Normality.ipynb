```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# First let's do an example with binomial distribution.
rand0 = np.random.binomial(20, 0.5, 100)
plt.hist(rand0)
plt.show()
```

```python
print(rand0.mean())
print(rand0.std())

plt.hist(rand0, bins=10, color='c')

plt.axvline(rand0.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(rand0.mean() + rand0.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rand0.mean() - rand0.std(), color='r', linestyle='dashed', linewidth=2)

plt.show() ;
```

Binomial distribution with small size of variables generated data with skew to the right of the mean value of 10.12 w/ sd at 2.66

```python
rand1 = np.random.chisquare(50, 100)
plt.hist(rand1)
plt.show() ;
```

```python
print('mean is', rand1.mean())
print('standard deviation is', rand1.std())
plt.hist(rand1, bins=10, color='g')
plt.axvline(rand1.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(rand1.mean() + rand1.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rand1.mean() - rand1.std(), color='r', linestyle='dashed', linewidth=2)
plt.show() ;
```

Chisquare distribution generates data that is not normally distributed, with skew to the left side of the histogram. Comparing values of mean and standard deviation, no usefull description can be provided using those two variables.

```python
rand2 = np.random.exponential(10, 100)
plt.hist(rand2)
plt.show() ;
```

```python
print('mean is', rand2.mean())
print('standard deviation is', rand2.std())

plt.hist(rand2, bins=10, color='y')
plt.axvline(rand2.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(rand2.mean() + rand2.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rand2.mean() - rand2.std(), color='r', linestyle='dashed', linewidth=2)
plt.show();
```

Exponential distribution generated non normally distributed variables with most values falling to the far left of the mean. Comparing values of mean and standard deviation, we have two closely matching values with mean at 10.27 and sd of 9.86

```python
rand3 = np.random.geometric(.2, 100)
plt.hist(rand3)
plt.show()
```

```python
print('mean is', rand3.mean())
print('standard deviation is', rand3.std())

plt.hist(rand3, bins=10, color='m')
plt.axvline(rand3.mean(), color='k', linestyle='solid', linewidth=2)
plt.axvline(rand3.mean() + rand3.std(), color='k', linestyle='dashed', linewidth=2)
plt.axvline(rand3.mean() - rand3.std(), color='k', linestyle='dashed', linewidth=2)

plt.show()
```

Geometric distribution has sd value very close to the mean, and is heavily skewed to the left, with most values falling on that side. Random values generated by geometric distribution are non normally distributed, with high standard deviatoin at 4.39, just bellow the mean value of 4.63.

```python
rand4 = np.random.logistic(0, 1, 100)
plt.hist(rand4)
plt.show()
```

```python
print('mean is', rand4.mean())
print('standard deviation is', rand4.std())

plt.hist(rand4, bins=10, color='b')
plt.axvline(rand4.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(rand4.mean() + rand4.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(rand4.mean() - rand4.std(), color='c', linestyle='dashed', linewidth=2)
plt.show()

```

Looking at logistic distribution we see that while most values are clustered around center, we have a skew on the left hand side with negative mean value of -0.0168 and standard deviation >1. 
This distribution did not generate normaly distributed data.

```python
rand5 = np.random.logseries(0.5, 100)
plt.hist(rand5)
plt.show()
```

```python
print('mean is', rand5.mean())
print('standard deviation is', rand5.std())

plt.hist(rand5, bins=5, color='r')
plt.axvline(rand5.mean(), color='k', linestyle='solid', linewidth=2)
plt.axvline(rand5.mean() + rand5.std(), color='k', linestyle='dashed', linewidth=2)
plt.axvline(rand5.mean() - rand5.std(), color='k', linestyle='dashed', linewidth=2)
```

Above histogram is of log series distribution, with most data falling on the mean and remaining variables tapering off on the right. This distribution does not generate normally distributed variables and comparing standard deviation to the mean does not give significant insight as to how the data will be distributed.

```python
rand6 = np.random.power(100, 100)
plt.hist(rand6)
plt.show
```

```python
print('mean', rand6.mean())
print('standard deviation', rand6.std())

plt.hist(rand6, bins=10, color='k')
plt.axvline(rand6.mean(), color='c', linestyle='solid', linewidth=2)
plt.axvline(rand6.mean() + rand6.std(), color='c', linestyle='dashed', linewidth=2)
plt.axvline(rand6.mean() - rand6.std(), color='c', linestyle='dashed', linewidth=2)
```

Here we see histogram of power distribution with most values on the right of the mean, with standard deviation value bellow the mean of 0.99. This shows that random variables generated by the power distribution will not be normally distributed.

Additionally:

    Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
    Add them together to create a third variable.
    Graph the third variable using a histogram.
    Compute the mean and standard deviation and plot them as vertical lines on the histogram and evaluate the descriptive statistics against the data.


```python
var1 = np.random.normal(5, 0.5, 100)
var2 = np.random.normal(10, 1, 100)
var3 = var1 + var2

plt.hist(var3)
plt.show()
```

```python
print('mean is', var3.mean())
print('standard deviation is', var3.std())

plt.hist(var3, bins=10, color='c')
plt.axvline(var3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='k', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='k', linestyle='dashed', linewidth=2)
plt.show()
```

Based on the information above and statistical analysis of the normal variables' sum shows skew on the right hand side of the histogram. Data generated by variable 3, which is a sum of two normally distributed variables, is not normally distributed, with most values not clustered in the middle and more values falling on the right hand side, thus showing a skew in data.
